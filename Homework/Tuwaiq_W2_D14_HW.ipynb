## Transform and Apply

Thanks to:
https://www.skytowner.com/explore/difference_between_methods_apply_and_transform_for_groupby_in_pandas



```python
dfta = pd.DataFrame({"A":[2,5,4],"B":[10,100,8],"group":["a","a","b"]})
dfta
```

```python
def f(my_df):
    # return the maximum value (scalar) in the entire my_df for each group
    return my_df.max().max()

dfta.groupby("group").apply(f)   # returns a Series
```

```python
# my_col is a Series representing a single column of each group
def f(my_col):
    # maximum value (scalar) in column gets broadcasted to become a Series of the same length as my_col
    return my_col.max()

dfta.groupby("group").transform(f)   # returns a DataFrame
```

## Merge / Join

```python
# Perform an inner merge (join) between two DataFrames on the "year" column
# Only rows with matching "year" values in both DataFrames are included
merged_inner = pd.merge(df1, df2, on="year", how="inner")

# Display the merged DataFrame
merged_inner
```

```python
# Perform an outer merge (join) between two DataFrames on the "year" column
# Includes all rows from both DataFrames
# Missing values are filled with NaN where there is no match
merged_outer = pd.merge(df1, df2, on="year", how="outer")

# Display the merged DataFrame
merged_outer
```

## Rolling / Expanding

```python
# Calculate a rolling mean (moving average) with a window of 3 rows
# Each value is the mean of the current and previous 2 rows
df["rolling_mean"] = df["passengers"].rolling(window=3).mean()

# Calculate an expanding mean (cumulative mean) over all rows up to the current one
df["expanding_mean"] = df["passengers"].expanding().mean()

df
```

## Datetime

```python
df_date = pd.DataFrame({
    "Event": ["Start", "Checkpoint", "End"],
    "Date": ["2025-01-01", "2025-01-15", "2025-02-01"]
})

# Convert "Date" column to datetime
df_date["Date"] = pd.to_datetime(df_date["Date"])

# Extract year and month
df_date["Year"] = df_date["Date"].dt.year
df_date["Month"] = df_date["Date"].dt.month
print(df_date)
```
